```python
from fastai.text import Tokenizer
from fastai.core import partition_by_cores
```

```python
from gensim import corpora
from gensim.parsing.preprocessing import remove_stopwords
from collections import defaultdict
```

```python
import spacy
```

```python
nlp = spacy.load('en')
```

```python
print(nlp.Defaults.stop_words)
```

```python
def construct_corpora(df):
    assert 'news' in df.columns
    assert 'title' in df.columns
    symbol_list = ["”","“",'—','t_up',]
    stopwords = nlp.Defaults.stop_words | set(['’s','she','it','said','you','’ve','’ll','n’t',])
    df = df.dropna()
    df['content'] = df['title'] + df['news']
    #df = [' '.join(x.strip().split()) for x in df.values]
    df = [gensim.parsing.strip_punctuation2(doc) for doc in df.content.values]
    #df = [gensim.parsing.remove_stopwords(doc) for doc in df]
    print(len(df))
    processed_docs = Tokenizer().proc_all_mp(partition_by_cores(df))
    processed_docs = [[word for word in doc if word not in stopwords and word not in symbol_list and len(word)>2] for doc in processed_docs]
    print(len(processed_docs))
    return processed_docs
```

```python
processed_week = construct_corpora(week_data[['title','news']])
```

```python
import logging
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
dictionary = corpora.Dictionary(processed_week)
```

```python
dictionary.filter_extremes()
```

```python
corpus = [dictionary.doc2bow(text) for text in processed_week]
```

```python
len(dictionary)
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=50,
                                      update_every=1, chunksize=30000, passes=5)
```

```python
#lda.save('newsgroups_50_lda.model')
```

```python
import pyLDAvis.gensim as gensimvis
import pyLDAvis
```

```python
vis_data = gensimvis.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)
```

```python
predicted_topics = lda[corpus]
```

```python
predicted_topics
```

```python
with open('./lda_articles.csv', 'w') as f:
    for i, doc in enumerate(predicted_topics):
        f.write(str(int(np.max(doc,axis=0)[0])) + ',' +' '.join(processed_week[i]) + '\n')
```
